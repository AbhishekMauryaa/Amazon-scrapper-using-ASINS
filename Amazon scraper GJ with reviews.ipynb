```python
# Create a new Chrome session
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode

driver = webdriver.Chrome()
driver.get('https://www.amazon.in')
```

```python
# Define the function to generate Amazon URLs from ASINs
def generate_amazon_url(asin):
    base_url = 'https://www.amazon.in/dp/'
    return base_url + asin
# Generate product URLs
product_urls = [generate_amazon_url(asin) for asin in asins]

# Print the generated product URLs
for url in product_urls:
    print(url)
```

```python
from bs4 import BeautifulSoup
import time
from tqdm import tqdm  # Make sure to import or define tqdm and selenium webdriver correctly


def clean_text(text):
    """Removes Unicode Left-to-Right Mark and extra whitespace from the text."""
    return text.replace('\u200e', '').strip()
# Assuming driver setup and product_urls definition are done elsewhere
# e.g., from selenium import webdriver
# driver = webdriver.Chrome()

# Initialize lists to store data
asins = []
product_titles = []
landing_page_images = []
ratings = []
ratings_count = []
stocks=[]
prices = []
mrps = []
xpaths_list = []
ships_from_list = []
sold_by_list = []
events_deals_list = []
offers_titles_list = []
brand_usps_list = []
additional_services_list = []
variations = []
product_details = []
highlighted_product_details = []
bullet_points = []
whats_in_the_box = []
brand_positioning_and_details_list = []
product_usps = []
technical_details = []
all_additional_details = []
ratings_by_star = []
ratings_overall_tags_list=[]
ai_gen_reviews = []
ratings_by_features_list = []
installation_details = []
brand_story_list = []
placeholder_images_list = []


# Loop through each URL
for url in tqdm(product_urls):
    driver.get(url)
    time.sleep(1)  # Be mindful of the website's load time and terms of use
    html_data = BeautifulSoup(driver.page_source, 'html.parser')
    product = html_data.find('div', {'id': 'a-page'}) #a-container
    parsed_url = urlparse(url)
    # Split the path at '/' and filter out empty values
    path_parts = [part for part in parsed_url.path.split('/') if part]
    
    # The ASIN typically follows 'dp' or 'gp/product' in the path
    if len(path_parts) > 1 and (path_parts[0] in ['dp', 'gp'] or (path_parts[0] == 'gp' and path_parts[1] == 'product')):
        # ASIN is the part immediately following 'dp' or 'product'
        asin = path_parts[1] if path_parts[0] == 'dp' else path_parts[2]
        asins.append(asin)

    
    if product:
# Extract title
        title_element = product.find('span', {'class': 'a-size-large product-title-word-break'})
        product_titles.append(title_element.text.strip() if title_element else 'N/A')
        
# Landing 
# Find all div tags with the specified class 'imgTagWrapper'
        divs = product.find_all('div', class_='imgTagWrapper')
        
        # Extract image URLs and store them in the landing_page_images list
        for div in divs:
            img = div.find('img')  # Find the img tag within the div
            if img and 'src' in img.attrs:  # Check if the img tag exists and has a 'src' attribute
                landing_page_images.append(img['src'])
        
# Extract ratings
        rating_element = product.find('span', {'class': 'a-icon-alt'})
        ratings.append(rating_element.text.strip() if rating_element else 'No Rating')

# Extract ratings count
        count_rating_element = product.find('span', {'id': 'acrCustomerReviewText', 'class': 'a-size-base'})
        ratings_count.append(count_rating_element.text.strip() if count_rating_element else 'No Rating')

# Extract stock availability
        stock_element = product.find('span', class_='a-size-medium a-color-success')
        if stock_element and "Currently unavailable." in stock_element.text.strip():
            stocks.append("Currently unavailable.")
            # Append 'N/A' for price and MRP if the item is unavailable
            prices.append('N/A')
            mrps.append('N/A')
        else:
            stocks.append("In stock")
            # Proceed with extraction only if the item is in stock
        
# Extract price
            price_element = product.find('span', {'class': 'a-price-whole'})
            prices.append(price_element.text.strip() if price_element else 'N/A')
        
# Extract MRP
            mrp_element = product.find('span', {'class': 'a-price a-text-price'})
            # Ensure we also handle potential absence of the inner span correctly
            if mrp_element:
                inner_span = mrp_element.find('span', {'aria-hidden': 'true'})
                mrps.append(inner_span.text.strip().replace('₹', '').strip() if inner_span else 'N/A')
            else:
                mrps.append('N/A')


# Category path extraction
        a_tags = product.find_all('a', class_='a-link-normal a-color-tertiary')
        xpaths = [a_tag.text.strip() for a_tag in a_tags]
        xpaths_str = '›'.join(xpaths)
        xpaths_list.append(xpaths_str if xpaths_str else 'N/A')

# Extract "ships from" information
        ships_from_element = product.find('span', {'class': 'a-size-small tabular-buybox-text-message'})
        ships_from_list.append(ships_from_element.text.strip() if ships_from_element else 'N/A')

# Extract "sold by" information
        sold_by_element = product.find('a', href=True, id='sellerProfileTriggerId')
        sold_by_list.append(sold_by_element.text.strip() if sold_by_element else 'N/A')

# Extract "events deals" information
        events_deals_element = product.find('span', {'id': 'dealBadgeSupportingText'})
        events_deals_list.append(events_deals_element.text.strip() if events_deals_element else 'N/A')

# Extract offers titles
        offers_elements = product.find_all('h6', {'class': 'a-size-base a-spacing-micro offers-items-title'})
        offers_titles = [element.text.strip() for element in offers_elements]
        offers_titles_str = ', '.join(offers_titles)
        offers_titles_list.append(offers_titles_str if offers_titles_str else 'N/A')

# Extract Brand USPs
        brand_usps_elements = product.find_all('span', {'class': 'a-size-small a-color-link a-text-normal'})
        brand_usps = [usp.text.strip() for usp in brand_usps_elements]
        brand_usps_str = ', '.join(brand_usps)
        brand_usps_list.append(brand_usps_str if brand_usps_str else 'N/A')

# Extract additional services information (warranty and prices)
        warranty_links = product.find_all('a', class_='a-link-normal mbb__st a-declarative')
        services_info = []

        for link in warranty_links:
            warranty_text = link.text.strip()
            price_element = link.find_next_sibling('span', class_='a-color-price')
            price_text = price_element.text.strip() if price_element else "Price not available"
            services_info.append(f"Warranty: {warranty_text}, Price: {price_text}")

        additional_services_str = "; ".join(services_info)
        additional_services_list.append(additional_services_str if services_info else "No additional services information")


##product_details
        table = product.find('table', class_='a-normal a-spacing-micro')
        # Initialize a dictionary for this product's parsed data
        parsed_data = {}
        
        # If the table is found, process its rows
        if table:
            rows = table.find_all('tr')
            for row in rows:
                cells = row.find_all('td')
                if len(cells) == 2:
                    label_cell = cells[0].find('span', class_='a-text-bold')
                    value_cell = cells[1].find('span', class_='po-break-word')
                    
                    # Ensure both label_cell and value_cell are found before proceeding
                    if label_cell and value_cell:
                        label = label_cell.text.strip()
                        value = value_cell.text.strip()
                        parsed_data[label] = value
        
        # Append the product's parsed data to the product details list
        if parsed_data:  # Ensure we don't add empty dictionaries
            product_details.append(parsed_data)
        else:
            # Optionally handle products without tables or details
            product_details.append({"Info": "No details available"})

        
#highlighted_product_details   
 # Initialize a temporary list for this product's details
        parsed_data = []

        # Extracting details from the table
        for row in product.select('table.a-normal > tbody > tr'):
            for cell in row.select('td.a-span6'):
                # Extracting text only, ignoring images
                feature_name = cell.find('span', class_='a-text-bold').text.strip()
                feature_value = cell.find('span', class_='handle-overflow').find_next_sibling('span').text.strip()
                parsed_data.append((feature_name, feature_value))
        
        # Append the parsed data for this product to the highlighted product details list
        highlighted_product_details.append(parsed_data)

##bullet_points
        
        # Select features/bullet points from the product page
        features = product.select("#feature-bullets .a-unordered-list .a-list-item")

        # Extract the text for each feature and store in a temporary list
        product_bullet_points = [feature.text.strip() for feature in features]

        # Append the product's bullet points to the main list
        bullet_points.append(product_bullet_points)


##whats_in_the_box        
        # Find the list item that contains "What's in the Box" details
        list_item = product.find("li", class_="postpurchase-included-components-list-item")

        # Then, find the 'span' element within the found 'li' element
        whats_in_the_box_detail = list_item.find("span", class_="a-list-item").text.strip() if list_item else "Details not available"

        # Append the extracted text to the whats_in_the_box list
        whats_in_the_box.append(whats_in_the_box_detail)

##brand_positioning_and_details
        # Find all divs that contain brand positioning and details
        brand_positioning_and_details = product.find_all('div', {'class': 'a-section a-spacing-base brand-snapshot-flex-row'})

        # Initialize a temporary list for storing brand details for the current product
        temp_brand_details = []

        for brand_positioning_and_detail in brand_positioning_and_details:
            # Find the <p> tag within the current div and get its text content
            paragraph_text = brand_positioning_and_detail.find('p').text.strip() if brand_positioning_and_detail.find('p') else None
            if paragraph_text:
                temp_brand_details.append(paragraph_text)

        # Combine all brand details into a single string and append to the main list
        brand_positioning_and_details_list.append("; ".join(temp_brand_details) if temp_brand_details else "No brand details found")

        brand_Usps = product.find_all('span', class_='brand-snapshot-chip-text')

        # Extract the text content of each span and store in a temporary list
        temp_product_usps = [brand_Usp.text.strip() for brand_Usp in brand_Usps]

        # Append the product's USPs to the main list
        product_usps.append(temp_product_usps if temp_product_usps else ["No USPs found"])

        # Find the technical specifications table by its ID
        table = product.find('table', id='productDetails_techSpec_section_1')

        # Find the technical specifications table by its ID
        table = product.find('table', id='productDetails_techSpec_section_1')
        
        # Initialize a dictionary to hold the technical specifications for the current product
        current_product_details = {}
        
        if table:  # Check if the table exists
            # Iterate through each row in the table
            for row in table.find_all('tr'):
                # Extract and clean the text of the header (th) and value (td)
                header = clean_text(row.find('th').text) if row.find('th') else "Unknown"
                value = clean_text(row.find('td').text) if row.find('td') else "N/A"
                
                # Add the extracted and cleaned information to the dictionary
                current_product_details[header] = value
        
        # Append cleaned technical details to the technical_details list
        technical_details.append(current_product_details if current_product_details else {"Info": "No technical specifications available"})

##additional details
        # Find the table by ID 'productDetails_detailBullets_sections1'
        table_additional = product.find('table', id='productDetails_detailBullets_sections1')
        
        # Initialize a dictionary to hold the scraped data
        additional_details = {}
        
        # Check if the table exists
        if table_additional:
            # Iterate through each row in the table body
            for row in table_additional.find_all('tr'):
                # Extract the text of the first column (header) and the second column (value)
                header = row.find('th').text.strip()
                value = row.find('td').text.strip()
                
                # Add the extracted information to the dictionary
                additional_details[header] = value
        else:
            # If the section is not found, append 'N/A' for error handling
            additional_details['Info'] = 'Additional details not available'

        # Append the dictionary to the list to ensure its length is 1
        all_additional_details.append(additional_details)

##ratings bt star
        # Find the table by ID 'histogramTable' and then find all 'tr' elements within it
        rating_table = product.find('table', id='histogramTable')
        
        # Check if rating_table exists before proceeding
        if rating_table:
            rating_rows = rating_table.find_all('tr')
        
            # Check if rating_rows is not empty before iterating over it
            if rating_rows:
                # Initialize an empty list to store rating data
                rating_data = []
        
                # Iterate through each row to extract the star rating and percentage
                for row in rating_rows:
                    # Initialize a dictionary for this row's data
                    row_data = {}
        
                    # Extract the star rating (e.g., "5 star", "4 star", etc.) from the first 'td' element
                    star_rating = row.find('td', class_='aok-nowrap').text.strip()
        
                    # Extract the percentage (e.g., "54%", "19%", etc.) from the last 'td' element
                    percentage = row.find_all('td')[-1].text.strip()
        
                    # Add the extracted information to the dictionary for this row
                    row_data['Star Rating'] = star_rating
                    row_data['Percentage'] = percentage
        
                    # Append the dictionary to the list
                    rating_data.append(row_data)
        
                # Append the list of rating data to ratings_by_star
                ratings_by_star.append(rating_data)
            else:
                # Append a default rating data if rating_rows is empty
                ratings_by_star.append([{'Star Rating': 'N/A', 'Percentage': 'N/A'}])
        else:
            # Append a default rating data if rating_table is not found
            ratings_by_star.append([{'Star Rating': 'N/A', 'Percentage': 'N/A'}])




##ratings overall tags
        positive_aspects = set([button.span.text for button in  product.find_all('button', class_='_cr-product-insights_style_positive-button__2jm-N')])
        mixed_aspects = set([button.span.text for button in product.find_all('button', class_='_cr-product-insights_style_mixed-button__yur1e')])
        
        # Combine positive and mixed aspects and convert to list
        all_aspects = list(positive_aspects.union(mixed_aspects))
        
        # Check if there are any aspects found
        if all_aspects:
            # If at least one aspect is present, append it to the list
            ratings_overall_tags_list.append(all_aspects)
        else:
            # If no aspects are found, append an empty string to maintain list length as one
            ratings_overall_tags_list.append([''])

 ##ai gen tag
        # AI-generated reviews logic correction
        p_element = product.find('p', class_='a-spacing-small')
        if p_element and p_element.span:
            ai_gen_review_text = p_element.span.text.strip()
        else:
            # Corrected logic to handle cases where the expected structure is missing
            ai_gen_review_text = 'AI-generated review not found'
        # Always append the result to the ai_gen_reviews list, ensuring order is maintained
        ai_gen_reviews.append(ai_gen_review_text)


##ratings by feature      
        temp_ratings_by_features = []

        sections = product.find_all("div", class_="a-section a-spacing-none")
        
        for section in sections:
            attribute_elem = section.find("span", class_="a-size-base a-color-base")
            rating_elem = section.find("span", class_="a-size-base a-color-tertiary")
            
            # Check if both attribute and rating elements are found
            if attribute_elem and rating_elem:
                attribute = attribute_elem.text.strip()
                rating = rating_elem.text.strip()
                
                # Create a new dictionary for each attribute and rating
                ratings_dict = {"Attribute": attribute, "Rating": rating}
                
                # Add the dictionary to the temporary list
                temp_ratings_by_features.append(ratings_dict)
    
        # After processing all sections for the current URL:
        if not temp_ratings_by_features:
            # If no ratings by feature were found, use a placeholder
            temp_ratings_by_features = [{"Attribute": "N/A", "Rating": "N/A"}]
    
        # Add the temporary list (wrapped in another list to ensure length 1) to the main list
        ratings_by_features_list.append([temp_ratings_by_features])


##installation details        
        # Find the span tag with id "ppdBundlesHeading" and extract the text
        heading_span = product.find('span', id='ppdBundlesHeading')
        heading_text = heading_span.get_text(strip=True) if heading_span else None
        
        # Find the span tag with id "ppdBundlesPriceValueId" and extract the text
        price_span = product.find('span', id='ppdBundlesPriceValueId')
        price_text = price_span.get_text(strip=True) if price_span else None
        
        # Check if both heading and price are present and add them to the list
        if heading_text and price_text:
            installation_details.append({
                "Heading": heading_text,
                "Price": price_text
            })
        
# ## variation 
        product_variations = []  # Temporarily store variations for the current product

        # Find all style type sections (Model, Size, etc.)
        style_sections = product.find_all('div', attrs={'class': 'a-section a-spacing-small'}, id=lambda x: x and x.startswith('variation_'))
        
        if not style_sections:
            product_variations.append({"variations": "na"})  # Use a dictionary for consistency
        
        for section in style_sections:
            style_type = section.find('label', class_='a-form-label').text.strip(': \n')
            options_titles = []

            list_items = section.find_all('li', title=True)
            for item in list_items:
                option_title = clean_text(item['title'].replace('Click to select ', ''))
                options_titles.append(option_title)

            product_variations.append({
                'style_type': style_type,
                'options': options_titles
            })

        variations.append(product_variations if product_variations else [{"variations": "na"}])



        

#brand story
        # Initialize an empty list for brand story content for each product
        brand_story_content = []

        # Attempt to find the main class containing the brand story elements
        main_class = product.find('div', class_='aplus-v2 desktop celwidget')

        if main_class:
            # Find all image tags within the main class
            images = main_class.find_all('img')
        
            # Extract image URLs and accompanying alt text, excluding GIFs
            for img in images:
                image_url = img.get('src')
                alt_text = img.get('alt', '')  # Default alt text to empty string if not present
                # Exclude GIF images
                if not image_url.endswith('.gif'):
                    brand_story_content.append({'type': 'image', 'url': image_url, 'alt_text': alt_text})
        
            # Find and clean all text within the main class
            text_sections = main_class.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'li'])  # Including more tags for broader text capture
            for section in text_sections:
                text_content = ' '.join(section.text.split())  # Clean the text by removing extra spaces and newlines
                if text_content:  # Only append if text_content is not empty
                    brand_story_content.append({'type': 'text', 'content': text_content})

        # Check if brand story content was found and append accordingly
        if not brand_story_content:
            # Append a placeholder if no content was found
            brand_story_content.append({'type': 'text', 'content': 'No brand story found'})

        # Append the brand story content or placeholder to the brand_story_list
        brand_story_list.append(brand_story_content)



#placeholder image
        # Assuming 'product' and 'placeholder_images_list' are already defined and initialized
        # Find the main class containing the images and text
        main_class = product.find('ul', class_='a-unordered-list a-nostyle a-button-list a-vertical a-spacing-top-micro gridAltImageViewLayoutIn1x7')
        
        if main_class:
            # Find all image tags within the main class
            images = main_class.find_all('img')
            
            # Initialize a list to store the placeholder image URLs
            placeholder_images = []
            
            # Extracting image URLs and accompanying text, excluding GIFs
            for img in images:
                image_url = img.get('src')
                alt_text = img.get('alt')
                # Exclude GIF images
                if not image_url.endswith('.gif'):
                    # Modify the URL to remove unnecessary parts
                    modified_url = re.sub(r'(\._SS|\SS)\d+_', '.', image_url)
                    placeholder_images.append(modified_url)
            
            # Append placeholder_images to placeholder_images_list
            placeholder_images_list.append(placeholder_images)
        else:
            print("Not able to extract placeholder images")
            
        # Ensure the list has a length of 1
        if len(placeholder_images_list) == 0:
            placeholder_images_list.append([])


        
    else:
        # If the product is not found, append 'N/A' or appropriate placeholders to all lists
        asins.append('N/A')
        
        ratings.append('No Rating')
        ratings_count.append('No Rating')
        stocks.append('N/A')
        prices.append('N/A')
        mrps.append('N/A')
        xpaths_list.append('N/A')
        ships_from_list.append('N/A')
        sold_by_list.append('N/A')
        events_deals_list.append('N/A')
        offers_titles_list.append('N/A')
        brand_usps_list.append('N/A')
        additional_services_list.append('N/A')
        #style_options_list.append('N/A')
        variations.append('N/A')
        product_details.append({'Info': 'No details available'})
        highlighted_product_details.append([])
        bullet_points.append([])
        whats_in_the_box.append('Details not available')
        brand_positioning_and_details_list.append('No brand details found')
        product_usps.append(['No USPs found'])
        technical_details.append({'No technical specifications available'})
        all_additional_details.append({'No additional details available'})
        ratings_by_star.append([])
        ratings_overall_tags_list.append([])
        ai_gen_reviews.append('AI-generated review not found ')
        ratings_by_features_list.append([])
        installation_details.append({'Heading': 'N/A', 'Price': 'N/A'})
        # brand_story.append([])  # Assuming an empty list for brand story if not found
        placeholder_images_list.append('N/A')

# Define lists to adjust
lists_to_adjust = [
    asins, product_titles, landing_page_images, ratings, ratings_count, stocks,prices, mrps, 
    xpaths_list, ships_from_list, sold_by_list, events_deals_list, offers_titles_list, 
    brand_usps_list, additional_services_list, variations, 
    product_details, highlighted_product_details, bullet_points, whats_in_the_box, 
    brand_positioning_and_details_list, product_usps, technical_details, all_additional_details, 
    ratings_by_star, ratings_overall_tags_list, ai_gen_reviews, ratings_by_features_list, 
    installation_details, brand_story_list, placeholder_images_list
]

num_urls = len(product_urls)
# Adjust list lengths
for data_list in lists_to_adjust:
    # Adjust list length if shorter than num_urls
    if len(data_list) < num_urls:
        data_list.extend(['N/A'] * (num_urls - len(data_list)))
    # Trim excess elements if longer than num_urls
    elif len(data_list) > num_urls:
        data_list = data_list[:num_urls]
```

```python
from bs4 import BeautifulSoup
import time

# Initialize lists to store data
names = []
ratings = []
rating_dates = []
titles = []
reviews_text = []
asins = []  # New list to store ASINs

# Loop through each URL in the product_urls list
for url in tqdm(product_urls):
    review_urls = []
    # Extract ASIN from the URL
    asin = url.split("/")[-1]  # Split URL by "/" and get the last part
    # Construct the review URL for the current product URL
    review_url = url.replace("dp", "product-reviews") + "/ref=cm_cr_arp_d_viewopt_fmt?formatType=current_format&sortBy=recent"
    review_urls.append(review_url)
    
    # Loop to iterate through each page of reviews
    i = 1
    while True:
        # Construct the URL for the current page of reviews
        page_url = review_url + "&pageNumber=" + str(i)
        
        # Get the HTML content of the page
        driver.get(page_url)
        time.sleep(1)
        html_data = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Extract review data from the page
        reviews = html_data.find_all('div', {'data-hook': 'review'})
        if not reviews:
            break  # Break the loop if no more reviews are found
        
        # Process each review on the current page
        for review in reviews:
            name = review.find('span', {'class': 'a-profile-name'}).text.strip()
            rating = review.find('span', {'class': 'a-icon-alt'}).text.strip()
            rating_date = review.find('span', {'data-hook': 'review-date'}).text.strip()
            title = review.find('a', {'data-hook': 'review-title'}).text.strip()
            review_text = review.find('span', {'data-hook': 'review-body'}).text.strip()
            
            # Append review data to the respective lists
            names.append(name)
            ratings.append(rating)
            rating_dates.append(rating_date)
            titles.append(title)
            reviews_text.append(review_text)
            asins.append(asin)  # Append ASIN to the list
        
        # Check for the next page of reviews
        next_page = html_data.find('li', {'class': 'a-last'})
        if not next_page:
            break  # Break the loop if there's no next page
        i += 1  # Increment page number

# Now you have the ASINs in the 'asins' list along with the other review data
# You can proceed to further process or analyze the collected review data
```
